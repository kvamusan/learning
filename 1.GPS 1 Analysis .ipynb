```python
import nltk
from textblob import TextBlob
import spacy

# Sample text for analysis
with open('gps1.txt', 'r', encoding='utf-8') as the_file:
    text = the_file.read()
    

# NLTK POS tagging
# nltk.download('averaged_perceptron_tagger')
tokens = nltk.word_tokenize(text)
pos_tags_nltk = nltk.pos_tag(tokens)

# SpaCy POS tagging
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
pos_tags_spacy = [(token.text, token.pos_) for token in doc]

# TextBlob POS tagging
blob = TextBlob(text)
pos_tags_blob = blob.tags

# Display POS tags from all three libraries
print("NLTK POS Tags:")
for word, pos_tag in pos_tags_nltk:
    print(f"{word}: {pos_tag}")

print("\nSpaCy POS Tags:")
for word, pos_tag in pos_tags_spacy:
    print(f"{word}: {pos_tag}")

print("\nTextBlob POS Tags:")
for word, pos_tag in pos_tags_blob:
    print(f"{word}: {pos_tag}")
```
