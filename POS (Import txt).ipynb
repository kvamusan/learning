```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

def identify_parts_of_speech(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    tokens = nltk.word_tokenize(text)
    pos_tags = nltk.pos_tag(tokens)
    return pos_tags

file_path = '1bar.txt' 
```

```python
print(identify_parts_of_speech(file_path))
```
