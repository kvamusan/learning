```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
with open('gps1.txt', 'r', encoding='utf-8') as the_file:
    text = the_file.read()
    

tokens = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(tokens)
# print(len(tokens))

pos_tags
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
print("POS Tags:")
for token in doc:
    print(f"{token.text}: {token.pos_}")
```
