```python
def identify_deixis(file_path):
    deictic_words = ['this', 'that', 'these', 'those', 'here', 'there', 'now', 'then', 'me', 'you', 'him', 'her', 'it', 'us', 'them']
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    words = text.split()
    deictic_in_text = [word for word in words if word.lower() in deictic_words]
    return deictic_in_text

# Replace '1bar.txt' with the path to your file
deictic_words_in_text = identify_deixis('1bar.txt')
print(deictic_words_in_text)
```

```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.corpus import treebank

def is_passive(tagged_sent):
    passive_voice = False
    for i in range(len(tagged_sent) - 1):
        if tagged_sent[i][1] == 'VBN' and (tagged_sent[i+1][1] == 'IN' or tagged_sent[i+1][1] == 'TO'):
            passive_voice = True
    return passive_voice

def identify_passive_sentences(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    sentences = nltk.sent_tokenize(text)
    passive_sentences = []
    for sentence in sentences:
        tagged_sent = pos_tag(word_tokenize(sentence))
        if is_passive(tagged_sent):
            passive_sentences.append(sentence)
    return passive_sentences

# Replace '1bar.txt' with the path to your file
passive_sentences_in_text = identify_passive_sentences("C:/Users/amusa/1bar.txt")
print(passive_sentences_in_text)
```
